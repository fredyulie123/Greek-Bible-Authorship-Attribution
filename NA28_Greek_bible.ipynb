```python
import re
import requests
from bs4 import BeautifulSoup
import time
import os
```

```python
Bookname = list(range(50,77))
Booknum = [28,16,24,21,28,16,16,13,6,6,4,4,5,3,6,4,3,1,13,5,5,3,5,1,1,1,22]
```

```python
data  = [] 

my_headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
AUTOTHROTTLE_ENABLED = True


for m in range(0,len(Bookname)):
    for k in range(0, Booknum[m]):
        url = 'https://www.academic-bible.com/en/online-bibles/novum-testamentum-graece-na-28/read-the-bible-text/bibel/text/lesen/stelle/'+str(Bookname[m])+'/'+str(k+1)+'0001/'+str(k+1)+'9999/ch'
        src = False
        for i in range(0,Booknum[m]): #error handlling
            try:
                response = requests.get(url, headers = my_headers)
                src = response.content
                break
            except:
                print('failed attempt #', i)
                time.sleep(2)
        if not src:
            print('Could not get page', url)
        else:
            print('Successfully get the page', url)


        soup = BeautifulSoup(src.decode('utf-8', 'ignore'),'lxml')
        
        title = soup.find('li',{'class':re.compile('scripture')})
        title_t = title.text.strip()
        contents = soup.findAll('p')
        for t in contents:
            try:
                vernum = t.find('span', {'class':re.compile('verse')})
                ver = vernum.text.strip()
                try:
                    fst = t.find('span',{'class':re.compile('greek')})
                    try:
                        fst1 = fst.text.strip()
                    except:
                        continue
                    texts = t.findAll('em')
                    allem = ''
                    for x in texts:
                        x = x.text.strip()
                        allem = allem + ' ' + x
                    if fst1 is not None:
                        alltext = fst1 + ' ' + allem
                    else:
                        alltext = allem
                    data.append([title_t,ver,alltext])
                except:
                    texts = t.find('span',{'class':re.compile('greek')})
                    txt = texts.text.strip()
                    data.append([title_t,ver,txt])
            except:
                try:
                    fst = t.find('span',{'class':re.compile('greek')})
                    try:
                        fst2 = fst.text.strip()
                    except:
                        continue
                    texts = t.findAll('em')
                    allem = ''
                    for x in texts:
                        x = x.text.strip()
                        allem = allem + ' ' + x
                    if fst1 is not None:
                        alltext = fst2 + ' ' + allem 
                    else:
                        alltext = allem
                    data.append([title_t,'nonvnum',alltext])
                    
                except:
                    texts = t.find('span',{'class':re.compile('greek')})
                    txt = texts.text.strip()
                    data.append([title_t,'novnum',txt])
               
            
       
```

```python
 with open ('NA28_Greek.csv', mode='w', encoding='utf-16') as f:
            for text in data:
                f.write(text[0] + '\t' + text[1] + '\t' + text[2] + '\n')
```
