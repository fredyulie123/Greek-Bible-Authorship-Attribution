```python
import re
import requests
from bs4 import BeautifulSoup
import time
import os
```

```python
Bookname = ['Matthew','Mark','Luke','John','Acts','Romans','1_Corinthians','2_Corinthians','Galatians','Ephesians','Philippians','Colossians','1_Thessalonians','2_Thessalonians','1_Timothy','2_Timothy','Titus','Philemon','Hebrews','James','1_Peter','2_Peter','1_John','2_John','3_John','Jude','Revelation']
Bookname1 = []
for x in Bookname:
    x = x.lower()
    Bookname1.append(x)
Bookname = Bookname1
Booknum = [28,16,24,21,28,16,16,13,6,6,4,4,5,3,6,4,3,1,13,5,5,3,5,1,1,1,22]
```

```python
data  = [] 

my_headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
AUTOTHROTTLE_ENABLED = True

for m in range(0,len(Bookname)):
    for k in range(1, Booknum[m]+1):
        url = 'https://biblehub.com/sbl/'+ Bookname[m] +'/'+ str(k) +'.htm'
        src = False
        for i in range(0,Booknum[m]): #error handlling
            try:
                response = requests.get(url, headers = my_headers)
                src = response.content
                break
            except:
                print('failed attempt #', i)
                time.sleep(2)
        if not src:
            print('Could not get page', url)
        else:
            print('Successfully get the page', url)
            
        soup = BeautifulSoup(src.decode('utf-8', 'ignore'),'lxml') 
        header = soup.find('div',{'id':re.compile('topheading')})
        header = header.text.strip()
        vernum = soup.findAll('span',{'class':re.compile('reftext')})
        content = soup.findAll('span',{'class':re.compile('greek')})
        for n in range(0,len(content)):
            verse_n = vernum[n].text.strip()
            verse_c = content[n].text.strip()
            data.append([header,verse_n,verse_c])
        
        
            
```

```python
## save as csv file
with open ('SLB_Greek-biblehub.csv', mode='w', encoding='utf-16') as f:
            for text in data:
                f.write(text[0] + '\t' + text[1] + '\t' + text[2] + '\n')
```
